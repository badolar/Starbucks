```python
driver = webdriver.Firefox()
starbucksurl=starbuckslinks1[0]
driver.get(starbucksurl)
page=driver.page_source
soup = BeautifulSoup(page, 'html.parser')
bootstrap=soup.find(name='div', attrs={'id':'bootstrapData'})
```

```python
bootstrap.text.split(',')  # This is a list
```

```python
bootlist=json.loads(bootstrap.text)
```
