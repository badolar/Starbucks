```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
from selenium import webdriver
from random import randint
from selenium.webdriver.common.keys import Keys
driver = webdriver.Firefox()
```

```python
starbucksurl='https://www.starbucks.com/store-locator'
# Turn tracker off on Firefox otherwise website will slow down
```

#### Find the search bar and type in the zip code

```python
zipcodes = pd.read_csv('../scraped_chain_locations/zip_codes.csv')
```

```python
y=40
a='link{}'.format(y)
a=[]
for x in range(16030,18000):  # x is zipcode index
    print(x)
    # Find the search bar
    driver.get(starbucksurl)
    time.sleep(1)
    search = driver.find_element_by_class_name("input___3eYjS")
    # Pull zip & location from the zipcode.csv, add 0 to 4 digit zip codes
    ziplocation = str(zipcodes.Zipcode[x])
    if len(ziplocation)==4:
        ziplocation = '0' + ziplocation
    location = (ziplocation +' '+ zipcodes.loc[x]['City'] +' '+ zipcodes.loc[x]['State'])
    search.send_keys(location)
    search.send_keys(Keys.RETURN)
    time.sleep(randint(3,5))
    page=driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    try:
        for entry in soup.find_all(name='a', attrs={'class':'relative pl5 color-textBlackSoft'}):
            a.append("https://www.starbucks.com" + entry['href'])
    except AttributeError:
        continue
```

```python
length = len(a)
print(length, '=number of items in link {}'.format(y))
```

```python
c='starbuckslinks{}'.format(y)
c=pd.Series(a)
```

```python
c= c.drop_duplicates().reset_index(drop=True)
```
